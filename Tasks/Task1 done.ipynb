Firstly, if we want to use *R* in Google Colab we have to change the programming lenguage to *R*, by default Google Colab uses Python. In order to do it, we have to access the notebook settings and we will do this by clicking on Edit and then on Notebook Settings.

Then, on the pop up menu, we can see that by default Google Colab works with Python as its programming language, so we just have to click on it and change it to R and then save, as we can see on the picture below.

Another way we can implement using *R* in Google Colab without changing the programming language to *R*, like we have explained before, consists of importing the package *rmagic* using the following command.

```python
%load_ext rpy2.ipython
```

After executing the previous command, we can now work using *R* language. However, in order to do so, we have to write %%R at the beginning of every chunk of code we write, in order for it to work.

It works as follows: it tells Google Colab that the following chunk of code is in *R* language, and because we have ran the previous command, it does not give us an error, but it rather runs our *R* code effortlessly. For example,

```python
%%R
x = seq(1, 20, 2)
x
```

Moreover, if we only want to run a line of *R* code and not a whole chunk of code, we just have to write %R at the beginning of it. For example,


```python
%R x = seq(1, 20, 2); x
```
